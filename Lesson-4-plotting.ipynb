## Line plots

We can make a line plot of one column against the other using the following syntax:

```
df.plot(x='x_values_column', y='y_values_column')
```
Example:

````python
# change in glacial volume per year in Europe

df.plot(x= 'year', y='europe')
````

Basic customization specifying the arguments in `plot()`

````python
# change in glacial volume per year in Europe

df.plot(x= 'year', 
        y='europe',
        title = 'Change in glacial volume per year in Europe',
       xlabel = 'Year',
       ylabel ='Change in glacial volume (km3)',
       color = 'green')
````

# Multiple line plots

We want to graph the change in glacial volume in the Arctic and Alaska. Update these arguments:
- 'y' : a list of column names that will be plotted against x
- 'color': specify the color of each column with a dictionary `{col1:'color1','col2':'color2'}`


````python
colors = {
    "arctic":'#F48Fb1',
    "alaska":"skyblue"
}

df.plot(x="year",
       y=["arctic", "alaska"],
       color=colors)
````

Create separate plots setting the `subset` argument to `TRUE`

````python
colors = {
    "arctic":'#F48Fb1',
    "alaska":"skyblue"
}

df.plot(x="year",
       y=["arctic", "alaska"],
       color=colors,
       subplots = True)
````

# `kind()` argument in `plot()`
parameter that controls what kind of plot we get is `kind`

## Scatter Plots

````python
penguins.plot(kind = 'scatter',
             x= 'flipper_length_mm',
             y='body_mass_g',
             alpha=0.4 # controls transparency points
             )
````

## Bar plot

Set `kind = bar` in the `plot()` method

Example: we want to know get data about the 10 penguins with the lowest body mass and visualize it
1. Find penguins with smallest body mass using `nsmallest()` method for series

````python
smallest = penguins.body_mass_g.nsmallest(10)
smallest
````

````python
# sort values in descending order
smallest = penguins.body_mass_g.nsmallest(10).sort_values(ascending=False)
smallest
````

````python
# plot smallest body masses as bars
smallest.plot(kind='bar')
````

What if we wanted to look at other data fo these smallest penguins.
We can use the `index` of the smallest series to select those rows in the original `penguins` data frame (use `loc`).

````python
penguins.loc[smallest.index]
````

# Histogram

To create a histogram, set `kind=hist` in `plot()`.

````python
# using plot without subsetting data - a mess again
penguins.plot(kind="hist")
````

````python
# distribution of flipper length measurements
# first select the data, then plot
penguins.flipper_length_mm.plot(kind="hist",
                               title="Flipper length distribution",
                               label="Flipper length (mm)",
                               color = "skyblue",
                               grid=True)
````
