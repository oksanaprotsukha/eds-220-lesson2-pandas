# Select columns
### Select a single column by a column name (the easiest method)

```
df['column_name']
```
This is the same syntax to access value from dictionary. Think of `pd.DataFrame` as a dictionary where the keys are the column names.

Example: we are interested in sea level rise

````python
# select a single column using []
annual_rise = df['annual_sea_level_rise']

# check the type of the output
type(df['annual_sea_level_rise'])
````

````python
annual_rise
````

`df['col_name']` is an example of selecting **by label**

**selecting by label** means we select data from the data frane using the names of the columns, not their position.

### Selecting a single columns with attribute syntax

**Syntax**: `df.column_name`

````python
# select annual seal level rise column
df.annual_sea_level_rise
type(df.annual_sea_level_rise)
````

## Selecting multiple columns using a list of column names
The syntax for selecting multiple columns using a list of columns names is:

```
df[ [`col1`], [`col2`], [`col3`],... [`colN`] ]
```

The list of columns_names `[`col1`], [`col2`], [`col3`],... [`colN`]` goes inside the selection brackets `[]`. That's why there are double brackets.

Notice each column name is a string.

Example: select change in glacial volume in Europe and Asia.

````python
# select columns with names "europe" and "asia"
europe_asia = df[['europe', 'asia']]

# print type of resulting selection
print(europe_asia)

# check the shape of the selection
print(europe_asia.shape)

# view the selection
europe_asia

# check type of europe_asia object
type(europe_asia)
````

## Select multiple columns using a slice

We will use `loc` (short for locate) selection. The general syntax for `loc` is 

```
df.loc[row_selection, column_selection]
```

where
- `row_selection` = rows we want to subset from `df`
- `column_selection` = columns we want to subset from `df`

Example: select change in glacial volume per year in all regions.

````python
df.head()
````

````python
# select all columns between europe and antarctica
df.loc[ : , 'europe':'antarctica'].head()
````

Notes about this:

- `row-selection` = `:` colon = means "select all the rows"
- `column_selection` = slice `'europe:antarctica'`

- we get both ends of the slice in the selected dataframe. Different from slicing in base Python and NumPy, where the endpoint is *not* included in the slice.

# Select rows

## Select rows using a condition

General syntax:

```
df[condition_on_rows]
```

**Example**: we are interested in data after 1996

````python
# select all rows with year > 1996
df[ df['year'] > 1996 ]
````

````python
# select all rows with year > 1996
df[ df.year > 1996 ]
````

Let's break this down:

- in the `df[conidtion_on_rows` syntax, we have that `condition_on_rows` = 'df['year']>1996'
- 'df['year']>1996' checks which rows return TRUE to the given condition, here year value is greater than 1996

````python
df['year']>1996
````

- `df['year']>1996` is a `pd.Series` with boolean values (`True/False`)
- `pd.Series` with boolean values (`True/False`) are often called **masks**
-  when  we pass `pd.Series` with masks to the selection brackets `.[]` we get the list of values that return True in the `pd.Series/mask`

Example: select data from 1970 to 1979 (including 1979)

````python
df [ df.year.isin(range(1970,1980)) ]
````

## Select rows using multiple conditions 

We can combine multiple conditions by surrounding each one in  parenthesis `()` and using the **OR operator** `|` and the **AND operator** `&`

The syntax:

```
# select rows of df that satisfy condition1 AND condition2
df[ (condition1) & (condition2)]
```
Example:

````python
# select rows with
# annual sea level rise < 0 mm OR annual sea level rise <0.8 mm

df[ (df['annual_sea_level_rise'] <0) | (df['annual_sea_level_rise']>0.8) ]
````

## Selecting rows by position

So sometimes we want to select rows on their actual *position* in the dataframe.
The syntax:

use `iloc` selection, sytax is:

```
df.iloc[row-indices]
```

`iloc` = stands for integer location based indexing.

````python
# select the fifth row = index 4

df.iloc[4]
````

````python
# select rows 23 through 30, including 30

df.iloc[23:31]
````

# Selecting rows and columns simultaneously 

We can do this using `loc` (for labels or conditions) or `iloc` (integer positions).

## Selecting rows and columns by label or conditions

Use `loc`:

```
df.loc[row-selection, column-selection]
```

`row-selection` and `column-selection` can be a condition or a subset of labels from the index or the column names.

**Example**:

````python
# select change in glacial volume in Europe per year after 2000

df.loc[ df.year > 2000, ['year','europe']]
````

## Selecting rows and columns by position

We use `iloc` with the syntax

```
df.iloc[row-indices, columns_indices]
```

Example

````python
# select rows with indices 3-7 (including 7) and the 4th and 5th columns

df.iloc[ 3:8, [3,4]]
````
